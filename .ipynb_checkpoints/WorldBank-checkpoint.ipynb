```python
import kagglehub
import pandas as pd
import numpy as np 
path = kagglehub.dataset_download("bhadramohit/world-bank-dataset")
```

```python
print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/world_bank_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import plotly.express as px
```

```python
fig = px.scatter(df, x='Population', y='Life Expectancy', trendline='ols')

fig.show()
```

```python
fig = px.scatter(df, x='GDP (USD)', y='Life Expectancy', trendline='ols')

fig.show()
```

```python
fig = px.scatter(df, x='GDP (USD)', y='Population', trendline='ols')

fig.show()
```

```python
fig = px.scatter(df, x='GDP (USD)', y='Access to Electricity (%)', trendline='ols')

fig.show()
```

```python
fig = px.scatter(df, x='GDP (USD)', y='Unemployment Rate (%)', trendline='ols')

fig.show()
```

```python
corr = df.corr(numeric_only=True)

fig = px.imshow(
    corr,
    text_auto=True,  
    color_continuous_scale='RdBu', 
    aspect='auto' 
)

fig.update_layout(title='Matriz de Correlação')
fig.show()
```
